```python
import pandas as pd
import requests

url_01 = "https://raw.githubusercontent.com/MK316/workspace/main/ASR01/data/rainbow_sentences.csv"
data1 = pd.read_csv(url_01)

url_02 = "https://raw.githubusercontent.com/MK316/workspace/main/ASR01/data/rainbow_wordfreq.csv"
data2 = pd.read_csv(url_02)
```

## 1) Generating English accented audio

```python
#@markdown ðŸš© Create audio with English accent for selected sentence
#@markdown Audio file will be saved as, e.g., "ES01.wav"
# !pip install gTTS
# from gtts import gTTS
# from IPython.display import Audio

sn = input('Which sentence? to audio? (1~19)')
sn = int(sn) - 1
txt = data1['Sentence'][sn]
print('S%d: %s'%((sn+1), txt))
filename = "ES" + str(sn+1)+".wav"

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en",
                    slow = False)
  
  gtts_object.save(filename)
  return Audio(filename)

etts(txt)
```

Generate 19 sentences to 19 audio files (ES1~ES19)

```python
#@markdown ðŸš© Create 19 audio files with English accent
#@markdown Audio file will be saved as, e.g., "ES1.wav"
# !pip install gTTS
# from gtts import gTTS
# from IPython.display import Audio

for i in range(0,19):
  txt = data1['Sentence'][i]
  print('S%d: %s'%((i+1), txt))
  filename = "ES" + str(i+1)+".wav"

  def etts(text):
    text_to_say = text

    gtts_object = gTTS(text = text_to_say,lang = "en",slow = False)

    gtts_object.save(filename)
  # return Audio(filename)

  etts(txt)
```
